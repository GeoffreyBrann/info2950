# Introduction  
Baseball is a sport that has been revolutionized by analytics. Introduced to pop culture by the film Moneyball, which depicted a small-market team (the Oakland Athletics) achieving success by using statistics in new and unique ways, analytics in baseball have advanced significantly since the Athletics in 2002. Not only are there different statistics to represent how good a player's results on the field are, new tecnhology and methods have been developed to further evaluate players based on their expected results. Here's an example: someone could hit the ball really hard directly at a fielder and make an out, and another player could hit the ball not as hard, but right between two fielders, and get a hit. One would prefer a hitter hit the ball hard rather than hitting it not as hard, but if you only look at results, the second player is better. Statistics like exit velocity (how hard the ball is hit) and launch angle (the angle at which the ball comes off the bat) measure inputs, rather than outcomes. These new statistics are called Statcast statistics, named after the technological system that makes their collection possible. This is a major shift in what the "best way to play" is: now that we can determine the best inputs, we no longer have to guess the best inputs to get the best outputs. Also, we can determine what the easiest way to get the best outputs are: you win games by scoring runs, and the easiest way to score runs is to hit home runs. Therefore, players are trying to hit as many home runs as possible.  
Our goal with this project was to determine "what combination of Statcast statistics best predicts the performance of a batter who qualified for the batting title in the years 2015-2019 in a single season". We had two more specific goals within this, namely finding the combination of Statcast statistics that best predicts a batter who qualified for the batting title in 2015-2019's offensive performance in terms of OPS+, wRC+, and xWOBA, and finding what combination of Statcast statistics best predicts a batter who qualified for the batting title in 2015-2019's UBR.  
Definitions:  

* OPS+:: On-base Plus Slugging Plus is the sum of On Base Percentage, the percentage of the time that a player gets on base due to a hit or a walk, and Slugging percentage (SLG),     which is total bases divided by at bats. Slugging percentage is a good measure of how much power a hitter has, because it values not just hits but the quality of hits, so a           better hit is worth more, and On Base Percentage is a good measure of a batter's general hitting skills and patience. OPS, the sum of those two percentages, gives a good              overall measure of a hitter. The Plus means that this stat is not a sum of percentages, but a number on the scale where 100 is average, 120 is 20% above average, etc. The             Plus also means that this statistic is adjusted both for ballpark and leaguewide performance, so not only can we more fairly compare players in the same league at the same            time, but we can also compare players across years more accurately.  
    
* wRC+:: Weighted Runs Created Plus uses a statistic called Runs to determine a player's performance. Teams score runs to try to win games, but these runs are scored as a team        effort. The Runs statistic is an effort to credit each player with a certain number of Runs that they created themselves, to evaluate them. The Plus does the same thing as            in OPS+, allowing for adjustment due to league-wide and ballpark run-scoring conditions, to make for a more fair comparison among players.  

* xWOBA:: Expected (x) Weighted On Base Average is the newest and most complex of these three statistics. It is a combination of Batting Average, On Base Percentage, and Slugging Percentage, but improves in areas in which each are lacking. Batting Average and On Base Percentage value each kind of hit the same, whether it be a single, double, triple, or home run. Slugging Percentage values these hits differently, but not correctly. It values bases, but is a double (2 bases) worth twice as much as a single (1 base)? OPS values one point of OBP and SLG as the same, but in reality one point of OBP is worth 1.8 of SLG. WOBA attempts to correctly account for each quality of hit and walks to give a better statistic to evaluate batters. WOBA is on the same scale as OPS, a decimal with 3 places. xWOBA is the same statistic as WOBA but with an added factor that takes some Statcast statistics like exit velocity, launch angle, and sprint speed to improve the existing WOBA statistic.  

* UBR:: Ultimate Base Running is a statistic that tracks how good of a base-runner a player is in units of Runs. By being an above-average baserunner, which means getting an above average number of bases during a certain event (hit, steal), you have a positive number of runs for this statistic. Below average, and you are negative, and average is 0.  

Statcast was first implemented in 2015, so despite baseball having records back to the late 1800s, we only have Statcast data since 2015. Secondly, "who qualified for the batting title" means that we only included batters with a sufficiently large number of plate appearences during the year, in an attempt to limit the problem of small sample sizes. Our last limitation, "in a single season", limits the scope of our project to not include a projection system for future years, because most existing projection models use 5 years of data to predict the next year's data, and we only have 5 years of data in total, so we could not determine how succesful our projection would be for a 6th year.  
We went about this by testing the significance of the diffrerence between the offensive results from different ranges of launch angles, the significance of the relationship between the variables of launch angle average (the average angle at which the ball left a player's bat over a year) and the three offensive performance statistics, as well as creating predictive models for statistics like UBR (a baserunning metric), Sprint Speed, and our offensive predictors (OPS+, wRC+, and xWOBA).  

We found significant differences between the offensive performance of different ranges of average launch angle over a season, and determined that the 4th quartile of launch angles had the best offensive performance and the 1st quartile of launch angles produced the worst offensive performance.  

We discovered that there were significance in the regression slopes between average launch angle and our offensive metrics, and from this could conclude that as launch angle goes up, offensive performance ameliorates.   
 
We created multiple models, then improved upon them using polynomial terms to predict offensive performance, UBR, and Sprint Speed from a small number of significant predictors. All of these models were satisfactory except for the UBR model, which gives us some insight into what does not make a player a good baserunner, but also some ideas as to what might make a player a good baserunner, like good instincts or intuition, which the Statcast systems cannot measure just yet. Of these models, our xWOBA model produced the most accurate results.  

The results of our analysis can be useful in a number of ways: it could be used to predict player performance, or determine what kind of prospect has more room to grow, and although our research may not be as useful for professional teams with their own analytics departments, it could definitely help someone's chances in a fantasy baseball league. It was exciting that with this publicly available data we could find such a wide breadth of significant results.    
Despite finding significant results, some weaknesses in our techniques or in the variables that we used inspired possible further research, such as in the case where we could develop improved regressions by using different launch angle metrics and different offensive predictors, or do even further analysis to improve our models.

# Preregistered Analysis: More In-Depth Analysis of Launch Angle
In our exploratory data analysis, we noticed that launch angle average did not have a high correlation coefficient with offensive value stats such as OPS+ and wRC+. We think this is because there is an ideal numerical range of average launch angles for peak offensive performance.
Over the past few years, there has been a major emphasis on changing the type of swing a batter would use to optimize their success, away from the traditional flat or even downwards-angled swing used to generate line drives, balls hit with a launch angle between 10 and 25 degrees, to a swing emphasizing flyballs, balls hit with a launch angle between 25 and 50 degrees (https://www.si.com/mlb/2018/03/21/evolution-swing-home-run-opening-day).  
To test our hypothesis that launch angle affects offensive performance, we divided all of the players' launch angles into different quartiles and determined that the 4th quartile of launch angles produced the best offensive results, the 2nd and 3rd quartiles could not be significantly differed from each other, and the first quartile produced the worst offensive performance.     
We also ran linear regressions on average launch angle and each of our three major offensive value stats. The goal was to see if there was a relationship between increasing launch angle and offensive performance. We found that despite low R-squared scores due to noise, the regression slopes were almost all significant.

```python
def launch_angle_avg_quartile(n):
    if(n >= q0 and n <= q1):
        return 1
    if(n >= q1 and n <=q2):
        return 2
    if(n >= q2 and n <= q3):
        return 3
    if(n >= q3 and n <= q4):
        return 4
```

```python
all_dfs = [data_2015,data_2016,data_2017,data_2018,data_2019]
all_data = pd.concat(all_dfs).reset_index(drop=True)

q0 = all_data['launch_angle_avg'].quantile(0)
q1 = all_data['launch_angle_avg'].quantile(0.25)
q2 = all_data['launch_angle_avg'].quantile(0.5)
q3 = all_data['launch_angle_avg'].quantile(0.75)
q4 = all_data['launch_angle_avg'].quantile(1)

dummy_all = map(launch_angle_avg_quartile, all_data['launch_angle_avg'])
all_data['launch_quartile'] = list(dummy_all)
```

```python
all_data_q1 = all_data[all_data['launch_quartile'] == 1]
all_data_q2 = all_data[all_data['launch_quartile'] == 2]
all_data_q3 = all_data[all_data['launch_quartile'] == 3]
all_data_q4 = all_data[all_data['launch_quartile'] == 4]

all_data_q1.nlargest(3, 'launch_angle_avg')
all_data_q2 = pd.merge(all_data_q1.nlargest(3, 'launch_angle_avg'), all_data_q2, how='outer')
all_data_q1 = all_data_q1.drop(all_data_q1.index[12])
all_data_q1 = all_data_q1.drop(all_data_q1.index[22])
all_data_q1 = all_data_q1.drop(all_data_q1.index[42])
all_data_q1 = all_data_q1.reset_index()
all_data_q2 = all_data_q2.reset_index()
```

However, to determine if these results are significant, we need to perform an unpaired t-test to see if the sets of offensive statistics have significant differences.

# Significance

An unpaired t-test is appropriate for determining if our results are significant because these quartiles are normal distributions, although they are quartiles of a larger sample the individual numbers were sampled independently, and the variances are mostly close (see below).  
The purpose of this t-test is to determine if the quartiles are significantly different to disprove the null hypothesis, that no quartile of average launch angle is better than any other at producing better offensive statistics. If we get sufficiently low p-values, (less than 0.05), the null hypothesis will be disproved and we can say that the launch angle quartile with the highest median, Q4, is best for producing the best offensive results.  

The first test we will conduct is a test to see if our variances are sufficiently close to run the t-test.

```python
print('Levene\'s Test for equality of variance. If the P=value is below 0.05, there is significant difference between the variances of the two sets')
print('wRC+ Q1 and Q2 P-Value::',stats.levene(all_data_q1['wRC_Plus'], all_data_q2['wRC_Plus'])[1])
print('wRC+ Q1 and Q3 P-Value::',stats.levene(all_data_q1['wRC_Plus'], all_data_q3['wRC_Plus'])[1])
print('OPS+ Q1 and Q2 P-Value::',stats.levene(all_data_q1['OPS_Plus'], all_data_q2['OPS_Plus'])[1])
print('OPS+ Q1 and Q3 P-Value::',stats.levene(all_data_q1['OPS_Plus'], all_data_q3['OPS_Plus'])[1])
print('OPS+ Q1 and Q4 P-Value::',stats.levene(all_data_q1['OPS_Plus'], all_data_q4['OPS_Plus'])[1])
print('xWOBA Q1 and Q2 P-Value::',stats.levene(all_data_q1['xwoba'], all_data_q2['xwoba'])[1])
print('xWOBA Q1 and Q3 P-Value::',stats.levene(all_data_q1['xwoba'], all_data_q3['xwoba'])[1])
```

Unfortunately, not all of the variances were close enough for us to perform t-tests on all combinations of quartiles within different offensive statistics using a normal independent t-test. Q1 had a smaller variance in each case, and therefore we cannot use a t-test and get a significant result if we compare any of the quartiles listed above. We can still compare wRC+ Q1 and Q4, and xWOBA Q1 and Q4, but other than that we cannot use Q1 in our tests. We can instead use a Welch's t-test, which allows us to still use an unpaired t-test even if the variances are significantly different. Everything else can be compared using our normal t-test.

```python
print('wRC+ unpaired t-test P-values')
print('Q1 and Q4::',stats.ttest_ind(all_data_q1['wRC_Plus'], all_data_q4['wRC_Plus'])[1])
print('Q2 and Q3::',stats.ttest_ind(all_data_q2['wRC_Plus'], all_data_q3['wRC_Plus'])[1])
print('Q2 and Q4::',stats.ttest_ind(all_data_q2['wRC_Plus'], all_data_q4['wRC_Plus'])[1])
print('Q3 and Q4::',stats.ttest_ind(all_data_q3['wRC_Plus'], all_data_q4['wRC_Plus'])[1])
print('\nOPS+ unpaired t-test P-values')
print('Q2 and Q3::',stats.ttest_ind(all_data_q2['OPS_Plus'], all_data_q3['OPS_Plus'])[1])
print('Q2 and Q4::',stats.ttest_ind(all_data_q2['OPS_Plus'], all_data_q4['OPS_Plus'])[1])
print('Q3 and Q4::',stats.ttest_ind(all_data_q3['OPS_Plus'], all_data_q4['OPS_Plus'])[1])
print('\nxWOBA unpaired t-test P-values')
print('Q1 and Q4::',stats.ttest_ind(all_data_q1['xwoba'], all_data_q4['xwoba'])[1])
print('Q2 and Q3::',stats.ttest_ind(all_data_q2['xwoba'], all_data_q3['xwoba'])[1])
print('Q2 and Q4::',stats.ttest_ind(all_data_q2['xwoba'], all_data_q4['xwoba'])[1])
print('Q3 and Q4::',stats.ttest_ind(all_data_q3['xwoba'], all_data_q4['xwoba'])[1])

print('\n\nwRC+ unpaired t-test P-values with different variances')
print('Q1 and Q2::',stats.ttest_ind(all_data_q1['wRC_Plus'], all_data_q2['wRC_Plus'], equal_var=False)[1])
print('Q1 and Q3::',stats.ttest_ind(all_data_q1['wRC_Plus'], all_data_q3['wRC_Plus'], equal_var=False)[1])
print('\nOPS+ unpaired t-test P-values with different variances')
print('Q1 and Q2::',stats.ttest_ind(all_data_q1['OPS_Plus'], all_data_q2['OPS_Plus'], equal_var=False)[1])
print('Q1 and Q3::',stats.ttest_ind(all_data_q1['OPS_Plus'], all_data_q3['OPS_Plus'], equal_var=False)[1])
print('Q1 and Q4::',stats.ttest_ind(all_data_q1['OPS_Plus'], all_data_q4['OPS_Plus'], equal_var=False)[1])
print('\nxWOBA unpaired t-test P-values with different variances')
print('Q1 and Q2::',stats.ttest_ind(all_data_q1['xwoba'], all_data_q2['xwoba'], equal_var=False)[1])
print('Q1 and Q3::',stats.ttest_ind(all_data_q1['xwoba'], all_data_q3['xwoba'], equal_var=False)[1])
```

From our t-test results, we can see that for most of our tests, the P-value is below 0.05 denoting significance. However, there were three tests that did not show significant difference, those being all three tests comparing Q2 and Q3. This is in line with what we observed before, with how similar the medians of those two quartiles were across all of our offensive statistics. Every other test showed significance, which means that we can say that they are significantly different sets of numbers and therefore can be compared.  
To compare them, I will compare each set's median and mean and then determine which quartile produces the best offensive results.  
One limitation of these tests, which we described before, is that a normal independent t-test requires that the variances be similar. However, because we did not have similar variances for some of our sets, we had to use a different t-test to find significance. Ideally, we would only use one kind of t-test when determining significance. Also, the t-test has fairly strict assumptions other than variance, which had to be fulfilled in order for us to use the test.

# Significance

To test the significance of the results found from this analysis, I will show the difference in R-Squared value and slope between the regressions with real values and regressions using fake values taken from normal distributions with the same mean, standard deviation, and sample size. I will then analyze the slope of the real regressions by showing possible slopes from randomizing the order of the independent variables but keeping the dependent variables constant, and then determining how unlikely it was to get the slopes that we did.  
If our real slopes are steeper than enough of the randomly generated slopes, we will conclude that these slopes are not random, and there is some relationship between our independent and dependent variables, average launch angle and offensive performance.

```python
def fake_regressions(n_samples, x_sample_mean, x_sample_sd, y_sample_mean, y_sample_sd, i, stat):
    regression_slopes = np.zeros(1000)
    regression_scores = np.zeros(1000)
    year = ""
    if(i == 5):
        year = 'All Years'
    else:
        year = str(i + 2015)
        
    for trial in range(1000):
        data = pd.DataFrame({'x': np.random.normal(x_sample_mean, x_sample_sd, size = n_samples),'y': np.random.normal(y_sample_mean, y_sample_sd, size = n_samples)})
        model = LinearRegression().fit(data[['x']], data[['y']])
        regression_slopes[trial] = model.coef_[0]
        regression_scores[trial] = model.score(data[['x']], data[['y']])
    
    print(year, stat, 'Large sample of random regressions\nRegression Slope: ', regression_slopes.mean())
    print('R-squared: ', regression_scores.mean())
    
def permute(y):
    # shuffle modifies the array itself, so first make a copy
    permuted_y = y.copy()
    # now randomize it
    np.random.shuffle(permuted_y)
    return permuted_y

def plot_model_line(model, x_extent, is_resampled=True):
    if is_resampled:
        color="0.8"
    else:
        color="blue"
    plt.plot(x_extent, x_extent * model.coef_[0] + model.intercept_, color=color)
    
def plot_fake_vs_real_and_significance(real_x_Series, real_y_Series, real_x_Dataframe, real_model, x_label, y_label, x_extent):
    plt.scatter(real_x_Series, real_y_Series)
    
    steeper_slopes = 0
    n_permutations = 10000
    permuted_slopes = np.zeros(n_permutations)

    for i in range(n_permutations):
        fake_y = permute(real_y_Series)
        fake_model = LinearRegression().fit(real_x_Dataframe, fake_y)
        plot_model_line(fake_model, x_extent)
        
        permuted_slopes[i] = fake_model.coef_[0]
    
        if np.abs(fake_model.coef_[0]) > np.abs(real_model.coef_[0]):
            steeper_slopes += 1
        
    plot_model_line(real_model, x_extent, is_resampled=False)
    print('Percent of permutated regression slopes steeper than the real regression slope: ',(steeper_slopes / n_permutations)*100,'%')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

# Conclusions  
From these analyses, we can conclude that there is a positive correlation between launch angle and offensive performance. However, the regression scores being relatively low casted some doubt as to whether or not our results were significant.  
After testing the significance of our resulting slopes, I would conclude that our results were significant, as with a 5% significance threshold only 1 of the 36 tests that we performed did not show significant results, and not by much. As said before, if we decrease our threshold to 3%, only 3 of the 36 become insignificant, and if we lower the threshold to 1%, 23 of the 36 are still significant.  
Despite our low R-Squared scores, our slopes were significant to the point that I believe there is a relationship between our offensive predictors and both launch angle average and launch quartile. Noise is to be expected in this kind of dataset, professional athletes are all elite in one way or another, and it would be unreasonable to say that only one value could prescribe success.  
One weakness of this analysis is that we used average launch angle as our predictor. Average launch angle is not a strong statistic because of the different kinds of ways an average can be composed: a player could hit only fly balls and only ground balls, and a player could hit only line drives, and both could end up with the same average launch angle despite completely different profiles.  
A better way to determine if trying to hit the ball with a certain launch angle might be to determine how succesful the shift in philosophy, from flat or downwards swings to uppercut swings, has been. We could do this by looking at flyball%, the percent of time that a player hits the ball with a launch angle between 25 and 50 degrees, and associating this with offensive stats, or even with a statistic like xISO, which measures a player's ability to hit for power while disregarding their general hitting ability.  
Despite its weaknesses, this analysis could be useful when determining a general strategy for evaluating hitters. Given a dead tie between two prospects, you might choose the one with more room for improvement by selecting the one with the lower average launch angle, in an attempt to alter their swing to better their offensive performance.
